## Import Package

Package digunakan untuk membantu dalam proses development. Package awal yang digunakan yaitu pandas dan numpy. Masing-masing digunakan untuk data analisis dan numerical computing.

## Memilih Fitur

Terdapat 4 fitur/atribut yang akan digunakan untuk pembuatan model klasifikasi dengan decision tree yaitu Sepal Length, Sepal Width, Petal Length dan Petal Width. Fitur yang telah dipilih akan dimasukan kedalam variabel X yang merupakan input variabel.
```python
X = iris[FEATURE_NAMES]
```
Untuk output target yaitu label/class yang akan diprediksi akan dimasukan kedalam variabel Y yang merupakan output variabel.
```python
y = iris['Class']
```

````python
import pandas as pd
import numpy as np

FEATURE_NAMES = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

iris = pd.read_csv('iris.csv')
X = iris[FEATURE_NAMES]
y = iris['Class']
````

Label masih berupa data kategorikal, sedangkan model machine learning bekerja dengan baik apabila tipe data berupa numerik. Dari hal ini maka label perlu diubah terlebih dahulu menjadi tipe data numerik.

# Membentuk Bagian Training dan Testing Model

Untuk dapat membuat model klasifikasi ML diperlukan pembagian data untuk proses training dan proses testing. Proses training merupakan proses untuk membentuk model tersebut dengan menggunakan data training. Proses testing merupakan proses untuk mengetes bagaimana performa dari model tersebut ketika diberikan data yang sebelumnya belum pernah dilihat pada proses training, yaitu dengan menggunakan data testing.
```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.35, random_state=6218013)
```

X_train dan y_train akan digunakan untuk proses training, lalu untuk X_test digunakan sebagai data testing untuk prediksi. Hasil prediksi tersebut akan dibandingkan dengan y_test. Pembagian datanya yaitu untuk data testing sebesar 35% dari jumlah data yang ada.

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.35, random_state=6218013)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape, y_test.shape)
````

# Import DecisionTree

Untuk dapat menggunakan algoritma DecisionTree maka perlu melakukan import DecisionTreeClassifier pada package sklearn terlebih dahulu.

````python
from sklearn.tree import DecisionTreeClassifier
````

## Melakukan Hyperparameter Tuning

Hyperparameter tuning merupakan proses mencari model yang terbaik dengan melakukan perubahan pada parameter model tersebut sebelum proses training. Terdapat 2 parameter yang dapat diubah yaitu criterion dan max_depth. Criterion merupakan parameter untuk mengukur ketidaksamaan (impurity) pembagian pada tree, sedangkan max_depth merupakan parameter yang mengatur seberapa dalam tree tersebut.

### Hyperparameter Tunning Dengan Criterion Entropy

Mengatur parameter criterion bernilai entropy serta max_depth masing-masing 3, 4, 5 dan 7

````python
modelEntropy1 = DecisionTreeClassifier(criterion="entropy", max_depth = 1)
modelEntropy1.fit(X_train,y_train)
````

````python
modelEntropy2 = DecisionTreeClassifier(criterion="entropy", max_depth = 7)
modelEntropy2.fit(X_train,y_train)
````

### Hyperparameter Tunning Dengan Criterion Gini

Mengatur parameter criterion bernilai gini serta max_depth masing-masing 3, 4, 5 dan 7

````python
modelGini1 = DecisionTreeClassifier(criterion="gini", max_depth = 1)
modelGini1.fit(X_train,y_train)
````

````python
modelGini2 = DecisionTreeClassifier(criterion="gini", max_depth = 7)
modelGini2.fit(X_train,y_train)
````

# Melakukan Prediksi

Setelah melakukan proses training, selanjutnya adalah melakukan prediksi. Untuk melakukan prediksi perlu menggunakan data testing, data testing yang digunakan adalah X_test. Hasil dari prediksi tersebut selanjutnya akan disimpan didalam variabel predictedTree

````python
predictedTree1 = modelEntropy1.predict(X_test)
````

````python
predictedTree2 = modelEntropy2.predict(X_test)
````

````python
predictedTree3 = modelGini1.predict(X_test)
````

````python
predictedTree4 = modelGini2.predict(X_test)
````

# Mengevaluasi Model

Dikarenakan decision tree merupakan klasifikasi multilabel maka untuk mengukur kinerja dari model yang telah dibuat adalah menggunakan matriks akurasi. Untuk caranya yaitu melakukan import pada package sklearn terlebih dahulu.

Secara sederhana hasil prediksi akan dibandingkan dengan hasil klasifikasi yang sesungguhnya. Maka dalam hal ini adalah variabel predictedTree akan dibandingkan dengan variabel y_test.

````python
from sklearn import metrics
metrics.accuracy_score(y_test, predictedTree1)
````

````python
metrics.accuracy_score(y_test, predictedTree2)
````

````python
metrics.accuracy_score(y_test, predictedTree3)
````

````python
metrics.accuracy_score(y_test, predictedTree4)
````

# Melihat Report Klasifikasi

Terdapat 3 matriks yang ada yaitu precision, recall dan f1-score. Secara sederhana semakin tinggi nilai dari ketiga matriks tersebut, maka semakin baik juga hasil klasifikasinya.

````python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictedTree1))
````

````python
print(classification_report(y_test, predictedTree2))
````

````python
print(classification_report(y_test, predictedTree3))
````

````python
print(classification_report(y_test, predictedTree4))
````
